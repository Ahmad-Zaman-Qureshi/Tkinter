```python
#Administrative Access
from tkinter import *
import mysql.connector
conn=mysql.connector.connect(host="localhost",user="root",password="12345",database="bus_station")
mycursor=conn.cursor()
window=Tk()
window.title("LOGIN")
import tkinter.messagebox
window.geometry("500x530")

#Menu
def menu():
    root1=tkinter.Tk()
    root1.geometry("500x350")
    root1.title("BMS")
    l1 = Label(root1,text = "Explore",font=("calibri",35,"bold"))
    l1.place(x=180,y=5)
    but = Button(root1,text = "Ticket Booth",width = "20",height = "2",bg = "black",fg = "white")
    but.place(x=190,y = 100)
    but = Button(root1,text = "Customer",width = "20",height = "2",bg = "black",fg = "white")
    but.place(x=190,y = 150)
    but = Button(root1,text = "Accounts",width = "20",height = "2",bg = "black",fg = "white")
    but.place(x=190,y = 200)
    but= tkinter.Button(root1, text="Exit",command=root1.destroy,bg='black',fg='white')
    but.place(x=250,y=250)
    root1.mainloop()        
    
#Exploration
class explore:
    def __init__(self,master):
        self.master=master
        self.label=Label(master,text="Administrative Access",width=25,fg="white",bg="black",relief="solid",font=("calibri",15,"bold")).pack(pady=10)
        
        self.label1=Label(master,text="Username",width=15,font=("calibri",14,"bold")).place(x=80,y=130)
        self.label2=Label(master,text="Password",width=15,font=("calibri",14,"bold")).place(x=79,y=170)
        
        self.var1=StringVar()
        self.var2=StringVar()
        
        self.entry1=Entry(master,textvar=self.var1).place(x=250,y=135)
        self.entry2=Entry(master,textvar=self.var2,show="*").place(x=250,y=175)
        
        self.button1=Button(master,text="Enter",command=self.check,fg="white",bg="black",font=("calibri",15,"bold")).place(x=230,y=335)
    def check(self):
        A1=self.var1.get()
        A2=self.var2.get()
        if A1=="project" and A2=="password":
            window.destroy()
            menu()
            print("Access Granted")
        else:
            tkinter.messagebox.showinfo("Alert","Invalid Username or Password")
            self.var1.set("")
            self.var2.set("")

obj=explore(window)
window.mainloop()
```

```python
#Customer
from tkinter import *
import mysql.connector
import tkinter.messagebox
conn=mysql.connector.connect(host="localhost",user="root",password="12345",database="bus_station")
mycursor=conn.cursor()

class Customer:
    def __init__(self,master,mycursor):
        self.master=master
        self.mycursor=mycursor
        master.title("Customer Details")
        
        self.mastlab=Label(master,text="Customer Entries",fg="white",bg="black",relief="solid",width="25",font=("calibri",15,"bold")).pack(pady=10)
        self.l1=Label(master,text="Entry Number",width=15,font=("calibri",14,"bold")).place(x=60,y=130)
        self.l2=Label(master,text="Customer's Name",width=15,font=("calibri",14,"bold")).place(x=60,y=170)
        self.l3=Label(master,text="Customer's Age",width=15,font=("calibri",14,"bold")).place(x=60,y=210)
        self.l4=Label(master,text="CNIC Number",width=15,font=("calibri",14,"bold")).place(x=60,y=250)
        self.l5=Label(master,text="Phone Number",width=15,font=("calibri",14,"bold")).place(x=60,y=290)
        self.l6=Label(master,text="Customer's Address",width=15,font=("calibri",14,"bold")).place(x=60,y=330)

        self.type1=IntVar()
        self.type2=StringVar()
        self.type3=IntVar()
        self.type4=StringVar()
        self.type5=StringVar()
        self.type6=StringVar()
        
        self.entry1=Entry(master,textvar=self.type1).place(x=250,y=135)
        self.entry2=Entry(master,textvar=self.type2).place(x=250,y=175)
        self.entry3=Entry(master,textvar=self.type3).place(x=250,y=215)
        self.entry4=Entry(master,textvar=self.type4).place(x=250,y=255)
        self.entry5=Entry(master,textvar=self.type5).place(x=250,y=295)
        self.entry6=Entry(master,textvar=self.type6).place(x=250,y=335)
        
        self.button=Button(master,text="Enter Data",command=self.add,fg="White",bg="black",font=("calibri",15,"bold")).place(x=230,y=390)
         
    def add(self):
        para1=self.type1.get()
        para2=self.type2.get()
        para3=self.type3.get()
        para4=self.type4.get()
        para5=self.type5.get()
        para6=self.type6.get()
        
        import mysql.connector
        db = mysql.connector.connect(host="localhost",user="root",password="12345",database="bus_station")
        mycursor = db.cursor()
        sql=("""CREATE TABLE IF NOT EXISTS Customer(Entry_Number int PRIMARY KEY AUTO_INCREMENT,Name varchar(30) NOT NULL,Age varchar(25) NOT NULL,CNIC varchar(30) NOT NULL,Phone varchar(20) NOT NULL,Address varchar(50) NOT NULL)""")
        mycursor.execute(sql)
        ins=("""INSERT INTO Customer(Entry_Number,Name,Age,CNIC,Phone,Address) VALUES("%d","%s","%d","%s","%s","%s")"""%(para1,para2,para3,para4,para5,para6))
        self.mycursor.execute(ins)
        tkinter.messagebox.showinfo("Alert","Data Insertion Succesful")
        conn.commit()
        
        self.type1.set("")
        self.type2.set("")
        self.type3.set("")
        self.type4.set("")
        self.type5.set("")
        self.type6.set("")
        
root=Tk()            
width = 520
height = 500
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
x = (screen_width/2) - (width/2)
y = (screen_height/2) - (height/2)
root.geometry("%dx%d+%d+%d" % (width, height, x, y))
root.resizable(0, 0)
obj=Customer(root,mycursor)      
root.mainloop()
```
